```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
fx = 100
Ax = 0.5

fc = 1000
Ac = 100

k = 500

fs = 50*fc

t = np.arange(0,2/fx,1/fs)
x = Ax*np.sin(2*np.pi*fx*t)
plt.plot(t,x)
plt.grid('minor')
plt.title('Sinal de entrada')
plt.ylabel('x(t)')
plt.xlabel('t')
```

Modulação AM

```python
c = Ac*np.sin(2*np.pi*fc*t)
plt.plot(t,c)
plt.grid('minor')
plt.title('Onda portadora')
plt.ylabel('p(t)')
plt.xlabel('t')
```

```python
xAM = np.multiply(x+3*Ax,c)
plt.plot(t,xAM)
plt.grid('minor')
plt.title('Sinal x(t) modulado em amplitude')
plt.ylabel('xAM(t)')
plt.xlabel('t')
```

Modulação FM

```python
f = 2*np.pi*(fc + 500*x)
theta = signal.lfilter(np.array([1]), np.array([1,-1]),f/fs)
xFM = Ac *np.sin(theta)
plt.plot(xFM)
plt.grid('minor')
plt.title('Sinal x(t) modulado em frequência')
plt.ylabel('xFM(t)')
plt.xlabel('t')
```

Modulação PM

```python
xPM = Ac*np.sin(2*np.pi*fc*t + x)
plt.plot(xPM)
plt.grid('minor')
plt.title('Sinal x(t) modulado em fase')
plt.ylabel('xPM(t)')
plt.xlabel('t')
```
